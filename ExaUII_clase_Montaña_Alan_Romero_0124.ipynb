```python
class Montana:
    def __init__(self, nombre, altura, cordillera):
        self.nombre = nombre
        self.altura = altura
        self.cordillera = cordillera

    def leerDatos(self):
        self.nombre = input("Ingrese el nombre de la montaña: ")
        self.altura = input("Ingrese la altura (msnm): ")
        self.cordillera = input("Ingrese la cordillera a la que pertenece: ")
        return self.nombre, self.altura, self.cordillera

    def mostrarDatos(self):
        print("Nombre de la montaña: ", self.nombre)
        print("Altura: ", self.altura)
        print("Cordillera: ", self.cordillera)
        return self.nombre, self.altura, self.cordillera

# Fin de la clase Montana

# Registro de la Montaña 1
print("Registro de Montaña numero 1")
montana1 = Montana("", "", "") # Creando objeto montana1
montana1.leerDatos()
montana1.mostrarDatos()
print("*-*-*-*-*-*-*-*-*-*-*-*-*-*-*--*-*-*-*")

# Registro de la Montaña 2
print("Registro de Montaña numero 2")
montana2 = Montana("", "", "") # Creando objeto montana2
montana2.leerDatos()
montana2.mostrarDatos()
print("*-*-*-*-*-*-*-*-*-*-*-*-*-*-*--*-*-*-*")

print("Software desarrollado por Alan Romero NL = 47")
```
